Rozwiązaniem powinien być pojedynczy plik `.py` oraz towarzyszący mu plik `.model`, które mają zawierać:

plik `NAZWISKO.py` ma zawierać klasę `ImagesClassifier`, która posiada:

* metodę `.predict(X)`, zwracającą klasę (w takiej postaci jak w pliku uczącym) dla każdego wiersza z macierzy `X`
* konstruktor jednoargumentowy `ImagesClassifier(f)`, ładującą model z pliku. Ma być jasne z kodu, że model ma formę sieci poznanej na zajęciach (lub ich zbioru).

W szczególności nie można:

* Używać sieci konwolucyjnych/neocognitronu/innych nie wprowadzonych architektur
* wykorzystywać wiedzy eksperckiej w postaci ręcznie zaprojektowanych cech które np. wykrywają uszy kota etc. ma to być sieć niezależna od typu zdjęć, czyli gdybyśmy podmienili któreś zdjęcia na zdjęcia innego zwierzęcia - wciąż powinna działać.

Można natomiast przeprowadzić ogólny preprocessing tj. zredukować liczbę wymiarów, zbinaryzować obrazek, uśrednić kolory etc.

Kod będzie testowany w następujący sposób:

    from NAZWISKO import ImagesClassifier

#### Jak będziemy oceniani?

Model zostanie oceniony na pewnym (nieznanym) zbiorze analogicznych obrazków. Innymi słowy istnieje przygotowany zbiór dodatkowych obrazów zwierząt/obiektów analogicznym formacie, które zostaną użyte by ocenić jak dobrze wasze modele **generalizują**.
